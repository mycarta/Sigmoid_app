#### Preliminaries

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Defining a sigmoid function

```python
def sigmoid(x,c,w):
    return 1/(1+np.exp(-(x-c)/w))
# creates a sigmoid function with parameters:
# c = the center of the sigmoid 
# w = the width of the sigmoid
```

#### Testing changing sigmoid width (& slope) - parameter b

```python
x = np.linspace(0,10,256)
y = sigmoid(x,5,1)
yw = sigmoid(x,5,1.5)
yn = sigmoid(x,5,0.5)
# Create the plot
plt.plot(x,y,lw=2,color='cyan')
plt.plot(x,yw,lw=2,color='green')
plt.plot(x,yn,lw=2,color='orange')
plt.show()
```

#### Testing shifting the center of the sigmoid - parameter a

```python
x = np.linspace(0,10,100)
y = sigmoid(x,5,0.5)
ya = sigmoid(x,3,0.5)
yb = sigmoid(x,7,0.5)
# Create the plot
plt.plot(x,y,lw=2,color='black')
plt.plot(x,ya,lw=2,color='red')
plt.plot(x,yb,lw=2,color='green')
plt.show()
```

#### Testing flipping of the sigmoid with left-right mirror

```python
x = np.linspace(0,10,100)
y = sigmoid(x,5,0.5)
yLR = sigmoid(x,5,-0.5)

# Create the plot
plt.plot(x,y,lw=2,color='blue')
plt.plot(x,yLR,lw=2,color='yellow')
plt.show()
```

#### Testing flipping of the sigmoid with up-down mirror

```python
x = np.linspace(0,10,100)
y = sigmoid(x,5,0.5)
yUD = -sigmoid(x,5,0.5)

# Create the plot
plt.plot(x,y,lw=2,color='green')
plt.plot(x,yUD,lw=2,color='magenta')
plt.show()
```

#### Testing negative sigmoid

```python
x1 = np.linspace(0,128,128) 
x2 = np.linspace(-128,0,128)
y1 = sigmoid(x1,64,12)
y2 = sigmoid(x2,-64,-12)

# Create the plot
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```
